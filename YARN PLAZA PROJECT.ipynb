```python
from bs4 import BeautifulSoup as bs #import beautiful soup library
import requests as rq #import request library
import pandas as pd #import pandas library
```

```python
url="https://www.yarnplaza.com/yarn/dmc"
result= rq.get(url) #send request to url
content= result.text
print(content)
```

```python
soup= bs(content, 'lxml')
```

```python
DMC= soup.find_all('h3',class_="productlist-title gtm-product-impression")#all DMC products
Natura_XL= [Natura.text.strip() for Natura in DMC][-1] #Natura_XL is the last DMC product
print(Natura_XL)
```

```python
prices= soup.find_all("span", class_="productlist-price")#all DMC product prices
price= [price.text.strip() for price in prices][-1]#price for Natura_X
pricelist= list(price) #rounding up the price to the nearest whole number
pricelist[2]= '9'
final1= pricelist[0] + pricelist[2]
print(final1)

```

```python
needle_sizes= soup2.find_all('tr') #obtaining needle size
needle_size= [needle.text.strip( ) for needle in needle_sizes][9]
needle_size[11:]
```
